```python
# Import necessary libraries
import os
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
filepath = ""
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        filepath = os.path.join(dirname, filename)
        
df = pd.read_csv(filepath)

# Check technical summary of the original dataset
df.info()
```

# Data Preparation

```python
# Convert timestamp column into datetime format
df["date"] = pd.to_datetime(df["time"], unit="s")

# Check columns' data types
df.dtypes
```

```python
# Check the oldest 5 data
df.head()
```

```python
# Check the latest 5 data
df.tail()
```

# Plot the Dataset

## Plot the closing price

```python
plt.figure(figsize=(14,4))

plt.plot(df["date"], df["close"], label="Close Price")

plt.title("BTC/USDT Daily Close Price")
plt.xlabel("Date")
plt.ylabel("Price (in USDT)")
plt.legend()
```

## Subplot all the prices

```python
df.plot(x='date', y=['open','close','low', 'high'], subplots=True, layout=(2,2), figsize=(16, 5))
```
